# Часть I. Теоретическая

Уравнение Бесселя имеет вид
$$ x^2y'' + xy' + (x^2 - \nu^2)y = 0,\\ \quad \nu \geq 0,  x > 0 $$

Решение имеет вид:

$$ y_1 = x^{\nu} \sum_{n=0}^{\infty}a_nx^n = \sum_{n=0}^{\infty}a_nx^{n+\nu}$$

Найдём соответствующие производные:

\begin{align}
& y_1' = \sum_{n=0}^{\infty} a_n(n+\nu)x^{n+\nu-1} \\
& y_1'' = \sum_{n=0}^{\infty} a_n(n+\nu)(n+\nu-1)x^{n+\nu-2}
\end{align}

Подставим их в исходное уравнение:
\begin{align}
&  x^2\sum_{n=0}^{\infty} a_n(n+\nu)(n+\nu-1)x^{n+\nu -2} + x\sum_{n=0}^{\infty}a_n(n+\nu)x^{n+\nu - 1} + (x^2 - \nu^2)\sum_{n=0}^{\infty}a_nx^{n+\nu}) = 0 \\
& \sum_{n=0}^{\infty} (a_n(n+\nu)(n+\nu-1)x^{n+\nu} + a_n(n+\nu)x^{n+\nu} + a_n(x^{n+\nu+2} - \nu^2x^{n+\nu})) = 0 \\
& \sum_{n=0}^{\infty} (a_nx^{n+\nu}((n+\nu)^2 - \nu^2) + a_nx^{n+\nu+2}) = 0 \\
\end{align}


Найдём рекуррент, идея в том, что если итоговая сумма равна нулю при любуых $x > 0$, значит коэффициент при каждом $x$ равен нулю.

\begin{align}
& x^{\nu+0}:  \ a_0(\nu^2 - \nu^2) = 0 \Rightarrow a_0 \in ℝ\\
& x^{\nu+1}: \space a_1(1 + 2\nu) = 0 \Rightarrow a_1 = 0 \\
& x^{\nu+2}: \space  a_2(4+4\nu) = -a_0 \Rightarrow a_2 = \frac{-a_0}{4+4\nu} \\
& x^{\nu+3}: \space a_3(9+6\nu) = a_1 = 0 \Rightarrow a_3 = 0 \\
& x^{\nu+4}: \space a_4(16+8\nu) = -a_2 = \frac{a_0}{4+4\nu} \Rightarrow a_4 = \frac{a_0}{(16+8\nu)(4+4\nu)}
\end{align}

Ну теперь можно сделать вывод, что:
\begin{align}
& a_{0} \in ℝ \\
& a_{1}=0 \\
& a_{2n}= \frac{(-1)^{n}a_0}{4^n n! (1+\nu) \dots (n+\nu)}, \ n\geq 2  \\
& a_{2n-1}=0, \ n\geq 2  \\
\end{align}


Решение может быть записано в виде:
$$ y_1(x) = a_0\sum_{n=0}^{\infty} \frac{(-1)^n}{4^n n! (1+\nu) \dots (n+\nu)}x^{2n+\nu}$$


Пользуясь [лекцией](https://www.youtube.com/watch?v=ne5oDOKSg6g), было несложно заметить, что можно сделать следующую замену:

$$a_0 = \frac{1}{4^\nu \Gamma(\nu + 1)}, \quad \Gamma(\nu) = \int_{0}^{+\infty} x^{\nu-1} e^{-x} dx$$

Тогда:
$$ y_1(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{\Gamma(n+1)\Gamma(n+\nu+1)}\left(\frac{x}{2}\right)^{2n+\nu} \triangleq J_{\nu}(x)$$


## Почему же это не общее решение???

Нетрудно заметить, что решением уравнения Бесселя будет линейное пространство решений размерности два, но мы нашли только один базисный вектор, поэтому это и не общее решение

йоуууу, как и было сказано выше, мы еще не нашли общее решение и пользуясь тем, что нам написали в условии ЛР знаем, что для общего решения надо найти еще одно решение в виде:

$$ y_2 = x^{-\nu} \sum_{n=0}^{\infty}b_nx^n = \sum_{n=0}^{\infty}b_nx^{n-\nu}$$

    
    

аналогично первому, для второго имеем:

$$ y_2 = b_0\sum_{n=0}^{\infty} \frac{(-1)^n}{4^n n! (1-\nu) \dots (n-\nu)}x^{2n-\nu} $$

В очередной раз обратимся к [лекции](https://www.youtube.com/watch?v=ne5oDOKSg6g) и положим коэффициент $b_0 = \frac{1}{4^{-\nu}Г(1-\nu)}$, имеем:

$$ y_2 = \sum_{n=0}^{\infty} \frac{(-1)^n}{Г(n+1)Г(n-\nu+1)}\left(\frac{x}{2}\right)^{2n-\nu} \triangleq J_{-\nu}(x)$$

теперь запишем общеее решение нашего уравнения при условии $2\nu ∉ ℤ$, так как в противном случае полученные решение ЛЗ и не состовляют ФСР (информация из [лекции](https://www.youtube.com/watch?v=ne5oDOKSg6g)):

$$ y_{о.р.} = c_1 \sum_{n=0}^{\infty} \frac{(-1)^n}{Г(n+1)Г(n+\nu+1)}\left(\frac{x}{2}\right)^{2n+\nu} + c_2\sum_{n=0}^{\infty} \frac{(-1)^n}{Г(n+1)Г(n-\nu+1)}\left(\frac{x}{2}\right)^{2n-\nu}$$

Найдем производную от общего решения:

$$ y'_{о.р.} = c_1 \sum_{n=0}^{\infty} \frac{(-1)^n(2n+\nu)}{Г(n+1)Г(n+\nu+1)}\left(\frac{x}{2}\right)^{2n+\nu -1} + c_2\sum_{n=0}^{\infty} \frac{(-1)^n(2n-\nu)}{Г(n+1)Г(n-\nu+1)}\left(\frac{x}{2}\right)^{2n-\nu -1}$$

$m$ - масса машины, которая уменьшается со скоростью $q$ сжигания топлива \

$\beta$ - коэффициент сопротивления воздуха \
$\mu$ - коэффициент трения между колёсами машины и землёй \
$u$ - постоянная скорость выбрасывания топлива относительно движения машины

$$m(0) = m_0, \quad v(0) = 0 \Rightarrow m(t) = m_0 - qt$$

Введём изменённую скорость V такую, что:

$$v(t) = \frac{m(t)\frac{dV}{dt}}{\beta V(t)}$$

Найдем производную от $v(t)$:


$$\frac{dv}{dt}  = \frac{\beta V\frac{dm}{dt}\frac{dV}{dt} + \beta mV\frac{d^2V}{dt^2} - \beta m\left(\frac{dV}{dt}\right)^2}{\beta^2 V^2} = \frac{-qV\frac{dV}{dt} + mV\frac{d^2V}{dt^2} - m\left(\frac{dV}{dt}\right)^2}{\beta V^2}$$

$$m \frac{-qV\frac{dV}{dt} + mV\frac{d^2V}{dt^2} - m\left(\frac{dV}{dt}\right)^2}{\beta V^2} + \frac{m^2(t)\frac{d^2V}{dt^2}}{\beta V^2(t)} + \mu mg - qu = 0$$
